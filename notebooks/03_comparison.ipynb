```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Model": ["Linear Regression", "Decision Tree", "XGBoost", "Deep Learning"],
    "Test_RMSE": [0.16343, 0.19878, 0.13984, 0.23998] 
                  }

df = pd.DataFrame(data)
df = df.sort_values(by="Test_RMSE")
df
```

```python
plt.figure(figsize=(8,5))

plt.bar(df["Model"], df["Test_RMSE"])
plt.title("Model Comparison (Test RMSE)")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=30)

plt.show()
```

```python
df.to_csv("results/final_results.csv", index=False)
```

## Conclusion

- XGBoost achieved the lowest RMSE, making it the best-performing model.
- Deep Learning did not outperform traditional ML models.
- This is likely due to the limited size of the dataset.
- Traditional ML models work better for structured/tabular data in this case.
